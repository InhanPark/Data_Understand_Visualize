```python
"""과학기술용 Library SciPy"""
# SciPY 는 주로 numpy array 를 기반으로 동작한다. 이 course 에서는 주로 SciPy.Stats 라는 통계용 sub-module 을 사용한다.
# 먼저 probability distribution function (확률분포함수), density and mass function, cumulative distribution function 의 용어를 알아야 한다.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# SciPy 에서 다양한 확률분포에 대해서 여러가지 값을 구하는 방법 (이런것이 있다는 것 정도만)

# Normal distribution (정규분포) 의 면적을 minus 무한대부터 0까지 구하는 방법
# Continuous variable 의 확률분포는 중앙값을 0 로 놓고 표현함.
# 따라서 이 경우는 0.5 가 나옴 (평균 이하일 확률은 0.5)
print(stats.norm.cdf(0))

# 스튜던트의 t-분포를 자유도 10 degree 로 median 구함 (정규분포 말고 t-분포 라는것도 있음)
print(stats.t(10).ppf(0.5))

# t-분포를 자유도 10 degree 로 97.5 percentile 구함 (97.5 위치 값)
print(stats.t(5).ppf(0.975))

# 표준 정규값이 3보다 같거나 작을 확률
print(stats.expon.cdf(3))

# 1에서의 표준정규분포밀도 함수의 높이
print(stats.norm.pdf(1))

# 동전을 10번 던져 앞면이 정확히 3번 나올 확률 (이런 경우를 binomial distribution 이항분포라고 함)
print(stats.binom(10, 0.5).pmf(3))

# 동전을 10번 던져 앞면이 3번 이하 나올 확률
print(stats.binom(10, 0.5).cdf(3))
```

```python
"""Plotting Library MatPlotLib"""
# Sine 함수를 그려보기

# MatPlotLib 는 기본적으로 x 와 y 숫자를 numpy array 로 만들어서 입력
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
print(type(x), x)
print(type(y), y)

# x, y array 를 주면 이것을 그림으로 그려줌
plt.plot(x, y)
```

```python
# plot 에 추가적인 내용을 표시함

plt.grid(True) # grid 를 그려줌
plt.plot(x, y, ":", lw=5, color="orange") # orage 점선으로 그려줌
plt.ylabel("Y", size=15)  # y 축 label 을 써 줌
plt.xlabel("X", size=15)  # x 축 label 을 써 줌
```

```python
# cosine data array 도 만들어 줌
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# sine 과 cosine curve 를 같이 그려줌
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
```

```python
# 두개의 plot 을 별도로 그려줌
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# height 2 and width 1 의 subplot 을 set up
# 첫번째 subplot 을 active 함
plt.subplot(2, 1, 1)

# 첫번째 plot 을 그림
plt.plot(x, y_sin)
plt.title('Sine')

# 두번째 plot 을 active 하고 그림
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')
```

```python
"""Matplotlib 보다 쓰기쉽고 그림이 예쁜 Seaborn"""
# seaborn 은 matplotlib 의 high-level interface 이다. Matplotlib 로 할 수 있는 대부분을 좀 더 쉽게 할 수 있다. Pandas 와 잘 맞는것 같음. 그러나 서로 장단점이 있어 같이 쓰는것 같음.

# Scatter plot 은 hostogram 과 같이 data 가 있으면 가장 먼저 그려보아야 하는것으로, x 와 y 사이의 관계를 파악하는데 사용한다. (x,y data 가 모두 있어야 함)

# .csv file data
url = "Cartwheeldata.csv"

# pandas Data Frame 으로 읽어들임
df = pd.read_csv(url)

# Scatterplot 을 plot (x 와 y 에 해당하는 column name 을 지정해 주면된다. Excel 보다 쉬움)
sns.scatterplot(x='Wingspan', y='CWDistance', data=df);
```

```python
# plot 에 legend (graph 의 '범례') 를 추가 
sns.scatterplot(x='Wingspan', y='CWDistance', hue="Gender", data=df);
```

```python
# categorical variable 로 x 를 변경한 경우의 plot. 같은 값에 해당하는 data 는 겹치지 않게 옆에 표시한다 (jittered). 이런것을 swarmplot 이라고 한다.
sns.swarmplot(x="Gender", y="CWDistance", data=df);
```

```python
# Boxplot 으로 mean(median), min, max, 25%, 75%, outlier 가 한번에 표시되어 있고, 'Age' 등 다섯가지 column 을 동시에 볼 수 있음. 얼핏보아도 'Score' 는 정규분포에 가까운 예쁜 모양의 histogram 일 것 같고, 'CWDistance 는 max 쪽 꼬리가 길고, 'Wingspan 은 min 쪽으로 약간 치우친 (skew) 모양임을 알 수 있음. 'Age' 는 고령자 두사람이 있고 ...
# Boxplot 은 내가 써 본 plot 중 가장 유용한 것 같음.

sns.boxplot(data=df.loc[:, ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

```python
# 'Female' 에 대해서만 plot 해 봄
sns.boxplot(data=df.loc[df['Gender'] == 'F', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

```python
# Male 에 대해서도 plot 해보고, Female data 와 비교해 볼 수 있음
sns.boxplot(data=df.loc[df['Gender'] == 'M', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
```

```python
# Boxplot 으로 대충의 data 를 훏어보고, 관심이 가는 column 은 histogram 으로 자세하게 분석함
sns.displot(df.CWDistance)
```

```python
# Count plot 은 data 의 수 (frequency) 를 확인할 수 있음
sns.countplot(x='Gender', data=df)
plt.xticks(rotation=-45)
```
