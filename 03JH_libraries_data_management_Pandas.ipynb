```python
"""Pandas 는 Excel 과 연계하여 이해하면 쉬움"""
# Pandas 는 Python Data Analysis Library 의 약어이다.
# Pandas 는 column 마다 data type 이 다를때 사용한다. Excel Sheet 와 같다.
# Pandas 에서는 주로 2가지 형태로 data 를 다룬다.
# - DataFrame : Excel sheet 처럼 2차원 data 표의 형태이다. 각 column 마다 data 가 다르다.
# - Series : DataFrame 에서 한 column 을 다룰때 많이 사용한다. 1차원 data 이다.
# Pandas 를 사용하려면 먼저 import 를 해주어야 한다.

import pandas as pd
```

```python
# 이번 Lab 에서 사용할 data 는 "Cartwheeldata.csv" 이다.
# '.csv' 는 'comma separated values' 의 약어로 아래와 같은 형태의 text file 이다. 각 data 들은 comma 로 구분되어 있다.

# ID,Age,Gender,GenderGroup,Glasses,GlassesGroup,Height,Wingspan,CWDistance,Complete,CompleteGroup,Score  <-- 첫번째 줄은 각 column 의 name 이 있다
# 1,56,F,1,Y,1,62,61,79,Y,1,7  <-- 두번째 줄 부터는 data 들이 있다.
# 2,26,F,1,Y,1,62,60,70,Y,1,8
# 3,33,F,1,Y,1,66,64,85,Y,1,7
# 4,39,F,1,N,0,64,63,87,Y,1,10
# 5,27,M,2,N,0,73,75,72,N,0,4

# csv file 을 'df' 라는 이름의 data frame 으로 읽어들인다.
fname = "Cartwheeldata.csv"
df = pd.read_csv(fname)

# 'df' 의 data type 은
print(type(df))
```

```python
# Data 를 읽어들인 후 분석을 하기 위해서는 먼저 data 를 확인하고 정리를 해야한다. 이 과정을 'Data Management' 라고 부른다. Pandas 는 기본적으로 Data Management 에 편리한 tool 이다.

# 읽어들인 data 의 앞부분 row 다섯개를 본다 (위의 .csv file 내용과 비교해보면 이해가 될 것이다.)
df.head()
```

```python
# column 의 이름을 보고 싶으면
df.columns
```

```python
# 각 column data 의 data type 을 보고 싶으면
df.dtypes
```

```python
"""DataFrame 에서 Cell 을 지정하기"""
# Excel 에서 특정구역에 있는 data 의 합을 구할때 '=SUM(C7:F9)' 처럼 명령을 준다. 즉 Excel 은 sheet 내의 특정 cell 의 위치를 가로방향(column) 은 alphabet, 세로방향(row)은 1에서 시작하는 숫자로 지정하고 Cell 이름은 '가로alphabet+세로번호' 로 매긴다. 그리고 영역 지정은 좌상단 cell 과 우하단 cell 을 지정하여 정한다. (Excel sheet 의 범위는 '1,048,576행 x 16,384열' 이다)
# Pandas (Numpy 도) DataFrame 의 최대 size 는 용량 기준으로 100GB 이다. 따라서 지정하는 방법이 다르다.

# 세로(row)를 먼저, 가로(column)를 나중에 지정한다. 그리고 다음과 같은 형식으로 범위로 지정을 한다.
# - [row 시작번호: row 끝번호+1, column 시작번호: column 끝번호+1] 
# - 시작번호는 0부터이다.
# - 에를들어 Excel 에서 'B3' cell 은 pandas/numpy 에서는 2:3, 1:2] 으로 표시된다. 아래 예 참조

a = df.iloc[2:3, 1:2]
print(a) 

# - 범위 지정시 숫자를 써주지 않으면 ':' 의 앞에서는 '0부터~' 뒤에서는 '끝까지~'라는 뜻이다.
```

```python
# cell 지정 방법만 익숙해지면 pandas 의 DataFrame 의 어느 data 나 사용할 수 있다. 예를들면 특정범위의 data 로 새로운 DataFrame 을 만들려면

sub_df = df.iloc[2:10, 1:2]
print(sub_df)
print(type(sub_df))
print()

# iloc 는 integer location 이라는 뜻이다. 즉 cell 위치를 숫자로 지정한다는 의미이다. Pandas 에서는 숫자말고 column 의 이름으로 지정하는 방법도 지원한다 (loc). 위의 예를 다시쓰면 이렇게 하면 된다.

sub_df = df.loc[2:10, "Age"]
print(sub_df)
print(type(sub_df))

# iloc 와 loc 에서 숫자로 지정한 범위의 끝번호가 달라짐을 주의한다. (iloc 가 1적다)

# 이를 가지고 DataFrame 의 data 를 사용하는 다양한 방법이 있다.
```

```python
# column name 을 'variable' 이라고 부르고 (attribute 등 다양한 이름이 있음), 각 row 는 observation 이라고 부른다 (record, object 등 다양한 이름이 있음)

# variable 'CWDistance' 의 모든 observation 을 봄
df.loc[:,"CWDistance"]
```

```python
# 이렇게 표시해도 같은 효과이다 (loc 안씀).
df["CWDistance"]

# 'df.CWDistance' 도 쓸 수 있지만, 문제가 생길 수 있다.
```

```python
# 여러개를 한번에 표시할 수도 있다.
df.loc[:,["CWDistance", "Height", "Wingspan"]]
```

```python
# 일정 구간의 row 만 지정할 수도 있다. (0~9 까지만)
df.loc[:9, ["CWDistance", "Height", "Wingspan"]]
```

```python
# column name 을 지정하지 않으면 전체를 보여준다.
df.loc[10:15]
```

```python
# iloc 로 지정하면 끝번호가 1이 줄어드는 것만 다르다.
df.iloc[1:5, 2:4]
```

```python
# iloc 로 지정하며 column name 을 사용하는 방법도 있다.
df.iloc[1:5, :][["Gender", "GenderGroup"]]
```

```python
"""Column 에 들어있는 Data 를 분석하기"""
# 'Gender' column 에 어떤 data 들이 들어있는지 확인한다. (잘못된것이 있는지 확인 가능)
df["Gender"].unique()
```

```python
# 'Gender' column 에 어떤 data 들이 들어있는지 확인한다. (잘못된것이 있는지 확인 가능)
df["GenderGroup"].unique()
```

```python
# 각 'Gender Group' 에 'Gender' 별로 몇명인지 확인한다. (Pivoting)
pd.crosstab(df["Gender"], df["GenderGroup"])
```

```python
# 이렇게 분석할수도 있다.
df.groupby(['Gender','GenderGroup']).size()
```
